```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('score.csv')
data.head()
```

```python
#Split the data
x=data.iloc[:,:-1].values
y=data.iloc[:,1].values
```

```python
# split the data into training set and testing set
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#fitting the linear model
model=LinearRegression().fit(x_train,y_train)
model
```

```python
#Plot the data
plt.scatter(x_train,y_train)
plt.plot(x_train,model.predict(x_train),color='red')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Percentage of Students based on Study Hours')
plt.show()
```

```python
model.score(x_train,y_train)
```

```python
#Plot the data
plt.scatter(x_test,y_test)
plt.plot(x_test,model.predict(x_test),color='red')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.title('Percentage of Students based on Study Hours')
plt.show()
```

```python
model.score(x_test,y_test)
```

```python
model.intercept_
```

```python
model.predict([[9.2]])
```

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
data=sns.load_dataset('iris')
```

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
label=LabelEncoder()
data['species']=label.fit_transform(data['species'])
data.head()
```

```python
x=data.iloc[:,0:4]
y=data.iloc[:,4:]
print(y.head())
x.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

```python
model=LinearRegression().fit(x_train,y_train)
model
```

```python
data.head()
```

```python
xtrain=model.predict(x_train)
xtrain.shape
```

```python
xtest=model.predict(x_test)
xtest.shape
```

```python
model.predict([[5.2,4,2,0.2]])
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
from sklearn.metrics import mean_squared_error
mean_squared_error(y_test,xtest)
```

```python
import seaborn as sns 
import pandas as pd
```

```python
data=sns.load_dataset('iris')
```

```python
data.shape
```

```python
data.describe()
```

```python
data.info()
```

```python
data.size
```

```python
data.head()
```

```python
x=data.iloc[:,0:4].values
y=data.iloc[:,4:].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=1)
print('x_train is : ', x_train.shape)
print('x_train is : ', y_train.shape)
print('x_train is : ', x_test.shape)
print('x_train is : ', y_test.shape)
```

```python
from sklearn import svm 
s=svm.SVC(kernel='linear')
s.fit(x_train,y_train)
s
```

```python
y_pred=s.predict(x_test)
y_pred
```

```python
s.score(x_train,y_train)
```

```python
s.score(x_test,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_pred,y_test)
```
